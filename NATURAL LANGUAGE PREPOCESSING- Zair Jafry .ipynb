```python
import nltk
nltk.download('punkt')
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize 
```

```python
text= word_tokenize(lst2str)
```

```python
print(text)
```

```python
data=[word for word in text if not word in stopwords.words()]
print(data)
```

```python
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemmmatizer=WordNetLemmatizer()
lem=[]
for x in data:
    lem.append(lemmmatizer.lemmatize(x))
```

```python
import re
data_update=[re.sub('[^a-zA-Z0-9]+','',_) for _ in lem]
```

```python
data_update
```

```python
# now forming clusters of the list lem[]
!pip install scikit-learn
```

```python
import sklearn
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.cluster import KMeans
from warnings import simplefilter 

simplefilter(action='ignore' , category=FutureWarning)
vectorizer= TfidfVectorizer(stop_words='english')
x=vectorizer.fit_transform(data_update)

k=5
model=KMeans(n_clusters=k,init='k-means++' ,max_iter=100 ,n_init=1)
model.fit(x)

print("Top per cluster ")
order_centroids=model.cluster_centers_.argsort()[:, ::-1] 
terms =vectorizer.get_feature_names()
for i in range(k):
    print("Cluster %d :" % i), 
    for ind in order_centroids[i, :10]:
        print('%s' % terms[ind])
    
                                        
    
```

```python
print('GUSSING THE STRING: \n')


Y = vectorizer.transform(["emoia"]) 
guessing = model.predict(Y) 
print(guessing)

```
